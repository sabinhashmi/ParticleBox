```python
ks=ROOT.TFile('/data3/tracking-upgrade-data/dump_comb/DD_ks_moore_big.root')
ksTree=ks.Get('Particle_Data')
l=ROOT.TFile('/data3/tracking-upgrade-data/dump_comb/DD_l_moore_big.root')
lTree=l.Get('Particle_Data')
```

```python
variables=['track_charge','track_chi2','track_chi2PerDoF','track_nLHCbIDs','track_p','track_phi','track_position_x',
           'track_position_y','track_position_r','track_pt','track_tx','track_ty','track_pseudoRapidity',
           'particle_hasScifi','particle_hasUT','particle_hasVelo','particle_isDown','particle_isDown_noVelo','particle_isLong',
           'particle_isLong_andUT','particle_isElectron']


features=['track_chi2PerDoF','track_nLHCbIDs','track_p','track_position_x',
           'track_position_y','track_position_r','track_pt','track_tx','track_ty','track_pseudoRapidity']


tracks=['particle_hasScifi','particle_hasUT','particle_hasVelo','particle_isDown','particle_isDown_noVelo','particle_isLong',
           'particle_isLong_andUT','partiole_isElectron']
```

```python
def plotFunKs(v,condition='track_p<100e3'):
    
    ksTree.SetLineColor(1) #black
    ksTree.Draw(v,'track_p<100e3')
    
    
    ksTree.SetLineColor(4) #blue
    ksTree.Draw(v,condition+'&'+'particle_isLong_andUT==1','SAME')

    ksTree.SetLineColor(2) #red
    ksTree.Draw(v,condition+'&'+'particle_isDown_noVelo==1','SAME')
    
    ksTree.SetLineColor(8) #green
    ksTree.Draw(v,condition+'&'+'particle_isElectron==1','SAME')
    
    
    ksTree.SetLineColor(6) #pink
    ksTree.Draw(v,condition+'&'+'particle_hasVelo==1','SAME')
    
    
    ksTree.SetLineColor(7) # lightblue
    ksTree.Draw(v,condition+'&'+'particle_isDown==1','SAME')
    
    
    

    
    
    
    
def plotFunLambda(v,condition='track_p<100e3'):
    
    lTree.SetLineColor(1)
    lTree.Draw(v,'track_p<100e3')
    
    
    lTree.SetLineColor(4)
    lTree.Draw(v,condition+'&'+'particle_isLong_andUT==1','SAME')

    lTree.SetLineColor(2)
    lTree.Draw(v,condition+'&'+'particle_isDown_noVelo==1','SAME')
    
    lTree.SetLineColor(8)
    lTree.Draw(v,condition+'&'+'particle_isElectron==1','SAME')
    
    
    lTree.SetLineColor(6)
    lTree.Draw(v,condition+'&'+'particle_hasVelo==1','SAME')
    
    
    lTree.SetLineColor(7)
    lTree.Draw(v,condition+'&'+'particle_isDown==1','SAME')
```

```python
canvas=ROOT.TCanvas('Canvas','Canvas',1500,3000)
legend=ROOT.TLegend()
canvas.Divide(2,5)
for i,v in enumerate(features):
    canvas.cd(i+1)
    plotFunKs(v)
canvas.Draw()
```
