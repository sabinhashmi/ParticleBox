```python
import ROOT
import uproot

import warnings
warnings.filterwarnings('ignore')

%jsroot on
```

```python
data=uproot.open('/home/hashmi/Files/DataFolder/ParticleBoxData/DownstreamData.root:Particle_Data')


data=data.arrays(library='pd')
data.drop('index',axis=1,inplace=True)
data=data[data['track_p']<100e3]


sample=data.sample(frac=0.001,random_state=46)
sample.head()
```

```python
%%time
canvas=ROOT.TCanvas()

legend=ROOT.TLegend(0.6,0.8,0.89,0.89)



downPlot = ROOT.TH1D('#Hits','#Hits',50,8.5,13.5)
ghostPlot = ROOT.TH1D('#Hits','#Hits',50,8.5,13.5)
for index,row in sample.iterrows():
    downPlot.Fill(row['track_nLHCbIDs'],row['Downstream']==True)
    ghostPlot.Fill(row['track_nLHCbIDs'],row['Downstream']==False)


ghostPlot.SetLineWidth(2)
ghostPlot.SetFillColor(9)
ghostPlot.Draw('HIST')

downPlot.SetLineWidth(2)
downPlot.SetFillColor(46)
downPlot.Draw('HIST  SAME')



downPlot.SetStats(0)
ghostPlot.SetStats(0)

legend.AddEntry(ghostPlot,'Ghost Tracks','f')
legend.AddEntry(downPlot,'True Tracks','f')


legend.Draw()

canvas.Draw()
# canvas.SaveAs('./Plots/Hits.eps')
```
