```python
import ROOT
```

```python
#open
Data=ROOT.TFile('/home/hashmi/Files/DataFolder/ParticleBoxData/DownstreamData.root')
data_tree=Data.Particle_Data
```

```python
def plot_fun(v,condition='track_p<100e3'):
    
    %jsroot on
    data_tree.SetLineColor(1)
    data_tree.Draw(v,'track_p<100e3')
    
    
    data_tree.SetLineColor(4)
    data_tree.Draw(v,condition+'&'+'Downstream==0','SAME')

    data_tree.SetLineColor(2)
    data_tree.Draw(v,condition+'&'+'Downstream==1','SAME')
```

```python
features=['track_chi2PerDoF','track_nLHCbIDs','track_p','track_position_x',
           'track_position_y','track_position_r','track_pt','track_tx','track_ty','track_pseudoRapidity']
```

```python
dataCanvas=ROOT.TCanvas('Canvas','Canvas',1500,3000)
# legend=ROOT.TLegend()
dataCanvas.Divide(2,5)
for i,v in enumerate(features):
    dataCanvas.cd(i+1)
    plot_fun(v)
dataCanvas.Draw()
```
